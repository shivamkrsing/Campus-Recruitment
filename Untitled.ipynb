```python
import numpy as np
def detect_outliers(data):
    outliers=[]
    threshold=3
    mean = np.mean(data)
    std =np.std(data)
    
    
    for i in data:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pt=detect_outliers(df2['Employability test'])
outlier_pt
```

```python
outlier_pt=detect_outliers(df2['mba_p'])
outlier_pt
```

```python
X=df2.drop("status",axis="columns")
y=df2.status
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(X_train,y_train)
mutual_info
```

```python
mutual=pd.Series(mutual_info)
mutual.index=X_train.columns
mutual.sort_values()
```

```python
from sklearn.feature_selection import SelectKBest
sel=SelectKBest(mutual_info_classif,k=10)
sel.fit(X_train,y_train)
X_train.columns[sel.get_support()]
```

```python
X_train=X_train[X_train.columns[sel.get_support()]]
```

```python
X_test=X_test[X_test.columns[sel.get_support()]]
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train,y_train)
reg.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn import metrics
preds = model.predict(X_test)
metrics.accuracy_score(y_test, preds)
```

```python
from sklearn.naive_bayes import GaussianNB
GNB=GaussianNB()
GNB.fit(X_train,y_train)
GNB.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LogisticRegression(), X, y, cv=cv)
```
